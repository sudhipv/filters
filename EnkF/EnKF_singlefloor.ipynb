### Gaussian model of shear frame building with 3 floors

### Observing 1 floor

```python
import numpy as np

import matplotlib.pyplot as plt

# First floor
u10 = 0.2
u11 = 0.05
u12 = 0.03


# Second floor
u20 = 0.3
u21 = 0.04
u22 = 0.02


# Third floor
u30 = 0.4
u31 = 0.06
u32 = 0.02


# np.rng(40)

np.random.seed(31)

nsamp =  1000

xi1 = np.random.normal(0,1, nsamp)
xi2 = np.random.normal(0,1, nsamp)

U1 = np.zeros((nsamp,1))
U2 = np.zeros((nsamp,1))
U3 = np.zeros((nsamp,1))


U1[:,0] = u10 + u11 *  xi1 + u12 *  xi2

U2[:,0] = u20 + u21 *  xi1 + u22 *  xi2

U3[:,0] = u30 + u31 *  xi1 + u32 *  xi2


P_k = np.zeros((3,3))

mu1 = np.mean(U1)
var1 = np.var(U1,ddof=1)

mu2 = np.mean(U2)
var2 = np.var(U2,ddof=1)

mu3 = np.mean(U3)
var3 = np.var(U3,ddof=1)


x_k_mean = np.array([mu1,mu2,mu3])

x_ki = np.array([U1[:,0],U2[:,0],U3[:,0]])

print("x_k_mean is", x_k_mean)

print(x_ki.shape)
# cov12 = np.cov(U1[:,0],U2[:,0])
# cov13 = np.cov(U1[:,0],U3[:,0])
# cov23 = np.cov(U2[:,0],U3[:,0])

# print(cov12)
# print(cov13)
# print(cov23)

data = np.stack((U1[:,0], U2[:,0], U3[:,0]), axis=0)

print(data.shape)

# print(U1[:,0].shape)

P_k = np.cov(data)

print("forecast cov is", P_k)

```

```python
### Ensemble of Data
from scipy.stats import norm

statesize = 3
observed = 1

x1 =  np.linspace(0.0,0.5, 100)

x2 =  np.linspace(0.0,0.5, 100)

x3 =  np.linspace(0.1,0.7, 100)


p1 = norm.pdf(x1, loc= mu1, scale = np.sqrt(P_k[0,0]))
p2 = norm.pdf(x2, loc= mu2, scale = np.sqrt(P_k[1,1]))
p3 = norm.pdf(x3, loc= mu3, scale = np.sqrt(P_k[2,2]))


H_k = np.zeros((observed,statesize))
H_k[0,0] = 1
print("H_k is",H_k)


np.random.seed(106)
# Noise strength
Gamma = 0.05


nsteps = 3

epsilon = np.random.normal(0,np.sqrt(Gamma),nsteps)

print(epsilon)

for i in range(0,nsteps):


    print("iteration step\n",i)

    d_k = H_k @ x_k_mean + epsilon[i]

    print("true data d_k is", d_k)


    e_i = np.random.normal(0,np.sqrt(Gamma),nsamp)

    print("e_i is", e_i.shape)

    d_ki = d_k + e_i

    print("d_ki is", d_ki)

    Gamma = np.var(e_i,ddof=1)

    print("sampled Noise strength is", Gamma)



    K_g = P_k @ H_k.T @ np.linalg.inv(H_k @ P_k @ H_k.T + Gamma)


    print("Kalman gain is \n", K_g)

    ### Update state using samples

    x_ai = np.zeros((statesize,nsamp))

    for i in range(nsamp):
        x_ai[:,i] = x_ki[:,i] + K_g @ (d_ki[i] - H_k @ x_ki[:,i])


    ### updated sample means and cov

    print(x_ai.shape)

    x_a_mean = np.mean(x_ai, axis=1)

    print("updated mean is", x_a_mean)

    P_a = np.cov(x_ai)

    P_a_check = (np.identity(statesize) - K_g @ H_k) @ P_k

    print("updated covariance is", P_a)

    print("updated covariance check is", P_a_check)


    p11 = norm.pdf(x1, loc= mu1, scale = np.sqrt(P_a[0,0]))
    p22 = norm.pdf(x2, loc= mu2, scale = np.sqrt(P_a[1,1]))
    p33 = norm.pdf(x3, loc= mu3, scale = np.sqrt(P_a[2,2]))


    plt.figure(i)
    plt.title("Floor 1")
    plt.plot(x1,p1,label = 'forecast')
    plt.plot(x1,p11,label = 'update')
    plt.legend()
    plt.show()

    plt.figure(i+1)
    plt.title("Floor 2")
    plt.plot(x2,p2,label = 'forecast')
    plt.plot(x2,p22,label = 'update')
    plt.legend()
    plt.show()

    plt.figure(i+2)
    plt.title("Floor 3")
    plt.plot(x3,p3,label = 'forecast')
    plt.plot(x3,p33,label = 'update')
    plt.legend()
    plt.show()

    x_k_mean = x_a_mean
    P_k = P_a



```
